### Pembersihan Data Teks

```python
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
from Sastrawi.Stemmer.StemmerFactory import StemmerFactory
nltk.download('punkt')
nltk.download('stopwords')
import re
```

Mengimport library yang akan digunakan nanti

```python
def tokenize_text(text):
    return word_tokenize(text)

df['Final_text'] = df['Text'].apply(tokenize_text)
df
```

Akan Melakukan tokenisasi dari text dan disimpan ke dalam Final_text. Bisa dilihat hasilnya dari tabel diatas

```python
print(df[:3])
```

Melihat apa saja yang tersipan di dalam df.

```python
def clean_text(tokens):
    cleaned_tokens = []

    for word in tokens:
        word = re.sub(r'[!$%^&*@#()_+|~=`{}\[\]%\-:";\'<>?]', '', word)
        cleaned_tokens.extend(word.split())
    
    cleaned_tokens = [word.lower() for word in cleaned_tokens]
    
    cleaned_tokens = [word for word in cleaned_tokens if not word.isdigit()]

    cleaned_tokens = [word for word in cleaned_tokens if "com" not in word]
    cleaned_tokens = [word for word in cleaned_tokens if "co" not in word]
    cleaned_tokens = [word for word in cleaned_tokens if "id" not in word]
    
    stop_words = set(stopwords.words('indonesian'))
    cleaned_tokens = [word for word in cleaned_tokens if word not in stop_words]

    cleaned_tokens = [word for word in cleaned_tokens if len(word) > 2]

    stemmer = StemmerFactory().create_stemmer()
    cleaned_tokens = [stemmer.stem(word) for word in cleaned_tokens]

    return cleaned_tokens

df['Final_text'] = df['Final_text'].apply(clean_text)

df
```

Akan dilakukannya pemberishan text dengan cara menghilangkan kata yang terdapat simbol-simbol special sepert $ atau #, menggantikan huruf besar menjadi huruf kecil, menghilangkan kata yang terdapat angka, menghilangkan kata yang terdapat .com dan .co.id di dalamnya, menghilangkan kata stopwords menggunakan stopword bahasa indonesia dari Sastrawi, menghilangkan kata yang hanya terdiri dari kurang dari 3 kata, dan menggantikan kata yang berimbuhan menjadi kata dasar. ## Text Representation

```python
import numpy as np
from sklearn.preprocessing import LabelEncoder
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report
from gensim.models import Word2Vec
```

Mengimport library yang akan digunakan nanti

### Embedding


```python
def average_word_vectors(words, model, vocabulary, num_features):
    feature_vector = np.zeros((num_features,), dtype="float64")
    n_words = 0
    for word in words:
        if word in vocabulary:
            n_words += 1
            feature_vector = np.add(feature_vector, model.wv[word])
    if n_words:
        feature_vector = np.divide(feature_vector, n_words)
    return feature_vector
```

menbuat function average_word_vectors untuk memprediksi nantinya

```python
tokenized_corpus = df['Final_text'].tolist()
model_w2v = Word2Vec(sentences = df['Final_text'], vector_size = 50, min_count = 3)
```

Megubah Final_text dalam df menjadi list

```python
X_w2v = [average_word_vectors(tokens, model_w2v, model_w2v.wv.index_to_key, 50) for tokens in tokenized_corpus]
X_w2v = pd.DataFrame(X_w2v)
```

Dengan menggunakan funtion average_word_vectors, akan dibuat dataframe baru X_w2v

```python
X_train_w2v, X_test_w2v, y_train, y_test = train_test_split(X_w2v, df['Label'], test_size=0.25, random_state = 42)
```

Akan membagi data menjadi 0,75 dan 0,25 untuk training dan testing data dengan x X_w2v dan y labelnya df.

```python
svm = SVC()
svmModel = svm.fit(X_train_w2v, y_train)
svmPred = svm.predict(X_test_w2v)
print(classification_report(y_test, svmPred))
```

Menggunakan SVM untuk memprediksi label. Dapat dilihat saat menggunakan SVM dengan metode text representation Word2Vec menghasilkan akurasi 25%

```python
randomForest = RandomForestClassifier()
RFmodel = randomForest.fit(X_train_w2v, y_train)
RFPred = randomForest.predict(X_test_w2v)
print(classification_report(y_test, RFPred))
```

Menggunakan RandomForest untuk memprediksi label. Dapat dilihat saat menggunakan RandomForest dengan metode text representation Word2Vec menghasilkan akurasi 25%

Akan melakukan prediksi label dengan salah satu text representation yaitu Word2Vec. Dengan dibandingkan SVM dan RandomForest, bisa dilihat kalau mereka menghasilkan akurasi yang sama yaitu 25%.
